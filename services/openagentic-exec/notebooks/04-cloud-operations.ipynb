````python
# Check available CLI tools and versions
import subprocess
import shutil

def get_version(cmd, version_flag='--version'):
    """Get version of a CLI tool"""
    if not shutil.which(cmd):
        return "Not installed"
    try:
        result = subprocess.run([cmd, version_flag], capture_output=True, text=True, timeout=10)
        output = result.stdout or result.stderr
        return output.strip().split('\n')[0][:80]
    except Exception as e:
        return f"Error: {e}"

print("Cloud CLI Tools Available:")
print("=" * 60)
print(f"kubectl:   {get_version('kubectl', 'version --client --short')}")
print(f"helm:      {get_version('helm')}")
print(f"aws:       {get_version('aws')}")
print(f"gcloud:    {get_version('gcloud')}")
print(f"az:        {get_version('az')}")
print(f"terraform: {get_version('terraform')}")
print(f"gh:        {get_version('gh')}")
````

## Kubernetes Operations

The cells below demonstrate common Kubernetes operations. These require a configured kubeconfig.

````python
# Kubernetes: Check cluster connection
import subprocess

def run_cmd(cmd):
    """Run a shell command and return output"""
    result = subprocess.run(cmd, shell=True, capture_output=True, text=True)
    if result.returncode != 0:
        return f"Error: {result.stderr}"
    return result.stdout

print("Kubernetes Cluster Info:")
print(run_cmd("kubectl cluster-info 2>/dev/null || echo 'Not connected to a cluster'"))
````

````python
# Kubernetes: List namespaces
print("Namespaces:")
print(run_cmd("kubectl get namespaces 2>/dev/null || echo 'Cannot list namespaces'"))
````

````python
# Kubernetes: List pods in default namespace
print("Pods in default namespace:")
print(run_cmd("kubectl get pods 2>/dev/null || echo 'Cannot list pods'"))
````

````python
# Helm: List installed releases
print("Helm Releases:")
print(run_cmd("helm list -A 2>/dev/null || echo 'Cannot list Helm releases'"))
````

## AWS Operations

AWS CLI operations require configured credentials (`~/.aws/credentials` or environment variables).

````python
# AWS: Check identity
print("AWS Identity:")
print(run_cmd("aws sts get-caller-identity 2>/dev/null || echo 'AWS not configured'"))
````

````python
# AWS: List S3 buckets (if configured)
print("S3 Buckets:")
print(run_cmd("aws s3 ls 2>/dev/null || echo 'Cannot list S3 buckets'"))
````

## GCP Operations

GCP CLI requires authentication (`gcloud auth login` or service account).

````python
# GCP: Check current project
print("GCP Configuration:")
print(run_cmd("gcloud config list 2>/dev/null || echo 'GCP not configured'"))
````

## Azure Operations

Azure CLI requires login (`az login`).

````python
# Azure: Check account
print("Azure Account:")
print(run_cmd("az account show 2>/dev/null || echo 'Azure not logged in'"))
````

## Infrastructure as Code Templates

The AI can generate IaC templates for you. Here's an example Kubernetes deployment:

````python
# Example: Generate a Kubernetes deployment YAML
deployment_yaml = """
apiVersion: apps/v1
kind: Deployment
metadata:
  name: my-app
  labels:
    app: my-app
spec:
  replicas: 3
  selector:
    matchLabels:
      app: my-app
  template:
    metadata:
      labels:
        app: my-app
    spec:
      containers:
      - name: my-app
        image: nginx:latest
        ports:
        - containerPort: 80
        resources:
          requests:
            memory: "128Mi"
            cpu: "100m"
          limits:
            memory: "256Mi"
            cpu: "200m"
"""

# Save to file
with open('deployment.yaml', 'w') as f:
    f.write(deployment_yaml)

print("Generated deployment.yaml:")
print(deployment_yaml)
print("\nTo apply: kubectl apply -f deployment.yaml")
````
