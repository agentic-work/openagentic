````python
# Import core libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set display options
pd.set_option('display.max_columns', 50)
pd.set_option('display.width', 200)

# Set visualization style
plt.style.use('seaborn-v0_8-whitegrid')
sns.set_palette('husl')

print(f"pandas version: {pd.__version__}")
print(f"numpy version: {np.__version__}")
````

````python
# Create sample data for demonstration
np.random.seed(42)

# Simulated system metrics data
dates = pd.date_range('2024-01-01', periods=100, freq='h')
data = {
    'timestamp': dates,
    'cpu_usage': np.random.normal(45, 15, 100).clip(0, 100),
    'memory_usage': np.random.normal(60, 10, 100).clip(0, 100),
    'disk_io': np.random.exponential(100, 100),
    'network_bytes': np.random.exponential(1000000, 100),
    'request_count': np.random.poisson(500, 100),
    'error_count': np.random.poisson(5, 100),
    'service': np.random.choice(['api', 'ui', 'worker', 'db'], 100)
}

df = pd.DataFrame(data)
print("Sample System Metrics Data:")
df.head(10)
````

````python
# Basic data exploration
print("Dataset Info:")
print(f"Shape: {df.shape}")
print(f"Columns: {list(df.columns)}")
print("\nData Types:")
print(df.dtypes)
print("\nSummary Statistics:")
df.describe()
````

````python
# Visualization: CPU and Memory Usage Over Time
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# CPU Usage over time
axes[0, 0].plot(df['timestamp'], df['cpu_usage'], color='#2ecc71', alpha=0.7)
axes[0, 0].fill_between(df['timestamp'], df['cpu_usage'], alpha=0.3, color='#2ecc71')
axes[0, 0].set_title('CPU Usage Over Time')
axes[0, 0].set_xlabel('Time')
axes[0, 0].set_ylabel('CPU %')
axes[0, 0].tick_params(axis='x', rotation=45)

# Memory Usage over time
axes[0, 1].plot(df['timestamp'], df['memory_usage'], color='#3498db', alpha=0.7)
axes[0, 1].fill_between(df['timestamp'], df['memory_usage'], alpha=0.3, color='#3498db')
axes[0, 1].set_title('Memory Usage Over Time')
axes[0, 1].set_xlabel('Time')
axes[0, 1].set_ylabel('Memory %')
axes[0, 1].tick_params(axis='x', rotation=45)

# Distribution of requests by service
service_requests = df.groupby('service')['request_count'].sum()
axes[1, 0].bar(service_requests.index, service_requests.values, color=['#e74c3c', '#9b59b6', '#f39c12', '#1abc9c'])
axes[1, 0].set_title('Total Requests by Service')
axes[1, 0].set_xlabel('Service')
axes[1, 0].set_ylabel('Request Count')

# Error rate by service
error_rate = df.groupby('service').apply(lambda x: x['error_count'].sum() / x['request_count'].sum() * 100)
axes[1, 1].bar(error_rate.index, error_rate.values, color=['#e74c3c', '#9b59b6', '#f39c12', '#1abc9c'])
axes[1, 1].set_title('Error Rate by Service (%)')
axes[1, 1].set_xlabel('Service')
axes[1, 1].set_ylabel('Error Rate %')

plt.tight_layout()
plt.savefig('system_metrics.png', dpi=150, bbox_inches='tight')
plt.show()
print("Chart saved to system_metrics.png")
````

````python
# Correlation analysis
numeric_cols = ['cpu_usage', 'memory_usage', 'disk_io', 'network_bytes', 'request_count', 'error_count']
correlation = df[numeric_cols].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation, annot=True, cmap='coolwarm', center=0, fmt='.2f',
            square=True, linewidths=0.5)
plt.title('Correlation Matrix - System Metrics')
plt.tight_layout()
plt.savefig('correlation_matrix.png', dpi=150, bbox_inches='tight')
plt.show()
print("Correlation matrix saved to correlation_matrix.png")
````
