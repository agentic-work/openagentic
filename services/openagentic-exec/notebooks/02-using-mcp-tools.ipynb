# Using MCP Tools in OpenAgentic

This notebook teaches you how to interact with MCP (Model Context Protocol) tools through the OpenAgentic platform.

## What are MCP Tools?

MCP tools are specialized capabilities that extend the AI's abilities:
- **Cloud Operations** - AWS, Azure, GCP management
- **Kubernetes** - Cluster management, Helm operations
- **Monitoring** - Prometheus, Loki, Alertmanager
- **Web** - Search, fetch, news aggregation
- **Admin** - Platform administration

## How to Use MCP Tools

Simply ask the AI in natural language! For example:
- "What pods are running in the default namespace?"
- "Show me the CPU usage for the last hour"
- "Search the web for Python best practices"

## Available MCP Tool Categories

| Category | Tools | Description |
|----------|-------|-------------|
| openagentic_kubernetes | 41 | K8s cluster management, Helm, pods, services |
| openagentic_web | 8 | Web search, fetch, news |
| openagentic_azure | 12 | Azure ARM, Graph, KeyVault, Cost |
| openagentic_gcp | 28 | GCP Compute, Storage, Vertex AI |
| openagentic_aws | 8 | AWS CLI, EC2, S3, Cost |
| openagentic_prometheus | 8 | Metrics querying |
| openagentic_loki | 9 | Log aggregation |
| openagentic_alertmanager | 9 | Alert management |

````python
# Example: Using the kubernetes Python client directly
# (The AI uses MCP tools, but you can also use the SDK)

from kubernetes import client, config

# Load kubeconfig (if available in your workspace)
try:
    config.load_incluster_config()  # When running in K8s
except:
    try:
        config.load_kube_config()  # When running locally
    except:
        print("No kubeconfig available - this is expected in demo mode")

# Example: List namespaces
try:
    v1 = client.CoreV1Api()
    namespaces = v1.list_namespace()
    print("Namespaces:")
    for ns in namespaces.items:
        print(f"  - {ns.metadata.name}")
except Exception as e:
    print(f"Cannot connect to K8s cluster: {e}")
````

````python
# Example: Using boto3 for AWS operations
import boto3

# Check if AWS credentials are configured
try:
    sts = boto3.client('sts')
    identity = sts.get_caller_identity()
    print(f"AWS Account: {identity['Account']}")
    print(f"AWS User ARN: {identity['Arn']}")
except Exception as e:
    print(f"AWS not configured: {e}")
    print("Ask the AI: 'Configure my AWS credentials'")
````

````python
# Example: Using httpx for web requests
import httpx

# The AI can use openagentic_web tools for more advanced searches
# Here's a simple example of making HTTP requests

response = httpx.get('https://httpbin.org/json')
print(f"Status: {response.status_code}")
print(f"Response: {response.json()}")
````
