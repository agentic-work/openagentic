Code Mode is an AI-powered development environment that combines:

## Quick Start

Run the cells below to explore your environment!

````python
# Check Python version
import sys
print(f"Python version: {sys.version}")
````

````python
# List available CLI tools
import subprocess

tools = ['kubectl', 'helm', 'aws', 'gcloud', 'az', 'terraform', 'git', 'go', 'python3', 'node']

print("Available CLI Tools:")
print("=" * 50)
for tool in tools:
    try:
        result = subprocess.run([tool, '--version'], capture_output=True, text=True, timeout=5)
        version = result.stdout.split('\n')[0] if result.stdout else result.stderr.split('\n')[0]
        print(f"  {tool}: {version[:60]}")
    except Exception as e:
        print(f"  {tool}: not available")
````

````python
# Check available Python packages
packages = ['pandas', 'numpy', 'matplotlib', 'seaborn', 'boto3', 'kubernetes', 'httpx']

print("Available Python Packages:")
print("=" * 50)
for pkg in packages:
    try:
        module = __import__(pkg)
        version = getattr(module, '__version__', 'installed')
        print(f"  {pkg}: {version}")
    except ImportError:
        print(f"  {pkg}: not installed")
````
